```python
import pandas as pd

anime_encoder_path = "../../data/anime_encoder.csv"
anime_csv_path = "../../data/anime.csv"
```

```python
item_encoder_df = pd.read_csv(anime_encoder_path)
item_encoder_df.rename(columns={
            "Orignal Id":"original_id",
            "Encoded Id":"encoded_id"
            }, inplace=True)
item_encoder_df
```
